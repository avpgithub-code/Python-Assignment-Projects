#### `READ THE BIG MART SALES DATA`

---

```python
big_mart_data = pd.read_csv('dataset/big_mart_sales.csv')
```

```python
big_mart_data.head()
```

#### Find out the number of unique `Item_Type` in each of the `Outlet_Identifier`

---

```python
# WRITE YOUR CODE HERE
pd.crosstab(big_mart_data['Outlet_Identifier'], big_mart_data['Item_Type'])
```

```python
# OR Another way to provide using GroupBy

num_of_unique_item_type_by_outlet = big_mart_data.groupby(['Outlet_Identifier','Item_Type']).count()['Item_Identifier']
num_of_unique_item_type_by_outlet
```

#### Find out the average `Item_Outlet_Sales` for each category of `Outlet_Size`

---

```python
# WRITE YOUR CODE HERE

avg_item_outlet_sales = big_mart_data.groupby(['Outlet_Size'])['Item_Outlet_Sales'].mean()
avg_item_outlet_sales
```
